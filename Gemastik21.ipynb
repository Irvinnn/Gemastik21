```python
import pandas as pd
import snscrape.modules.twitter as sntwitter
import itertools
```

```python
lokasi = ['Banten', 
          'Jakarta',
          'Jawa Barat', 
          'Jawa Tengah',
          'Yogyakarta']
```

```python
iter = 0
dic = {}

for x in lokasi:
    search = "since:2021-04-01 (butuh pandemi) near:'{}'".format(x)
    scraped_tweets = sntwitter.TwitterSearchScraper(search).get_items()
    dic[iter] = pd.DataFrame(scraped_tweets)[['user',
                                          'date', 
                                          'content']]
    dic[iter]['Location'] = x
    iter += 1
```

```python
dic
```

```python
result = dic[1].append([dic[2], dic[3],dic[4]])
result = result.reset_index()
```
